# ResNet50 implementation

```python
import torch.nn as nn
import torchvision.models as models
```

```python
# Loaing pretrained ResNet50 
class BreastCancerResNet(nn.Module):
    def __init__(self, num_classes=3):
        super(BreastCancerResNet, self).__init__()
        self.resnet = models.resnet50(pretrained=True)
        
        # Freeze all layers except the final classification layers
        for param in self.resnet.parameters():
            param.requires_grad = False
        
        # Replace the fully connected layer for 3-class classification
        self.resnet.fc = nn.Sequential(
            nn.Linear(self.resnet.fc.in_features, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        return self.resnet(x)

        
```

```python
# Instantiate model
model = BreastCancerResNet()
print(model)
```
